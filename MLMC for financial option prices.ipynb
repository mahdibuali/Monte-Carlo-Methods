MLMC for estimating Options price

```python
import numpy as np
l0 = 4
real = 0.11545579187590832
L = 9
T = 1
r = 0.05
var = 0.2
M = 100
N = [((L -l0 +1) * np.power(2, 2 * L - i)) for i in range(l0, L + 1)]
df = np.zeros((M, L-l0+1))
for k in range(M):
  value = 0
  for l in range(l0, L + 1):
    if l == l0:
      dl = 1/np.power(2, l)
      S = np.ones(N[l - l0])
      for i in range(T * np.power(2, l)):
        Z = np.random.normal(0, np.sqrt(dl), N[l - l0])
        S = S + r * dl * S + var * S * Z 
      P =  S - 1
      P[P < 0] = 0
      P = np.exp(-r) * P
      value += np.sum(P)/N[l - l0]
      # print(np.sum(P)/N[l - l0])
    else:
      dl1 = 1/np.power(2, l)
      dl2 = 1/np.power(2, (l-1))    
      S1 = np.ones(N[l - l0])
      S2 = np.ones(N[l - l0])
      for i in range(T * np.power(2, (l-1))):
        Z1 = np.random.normal(0, np.sqrt(dl1), N[l - l0])
        S1 = S1 + r * dl1 * S1 + var * S1 * Z1
        Z2 = np.random.normal(0, np.sqrt(dl1), N[l - l0])
        S1 = S1 + r * dl1 * S1 + var * S1 * Z2
        S2 = S2 + r * dl2 * S2 + var * S2 * (Z1 + Z2)
      P1 =  S1 - 1
      P1[P1 < 0] = 0
      P1 = np.exp(-r) * P1
      P2 =  S2 - 1
      P2[P2 < 0] = 0
      P2 = np.exp(-r) * P2
      temp = np.sum(P1 - P2)/N[l - l0]
      value += temp
      df[k, l-l0] = temp

print(value)
```

```python
import numpy as np
l0 = 0
real = 0.11545579187590832
real_as = 0.05802055815398626
real_look = 0.463271656263829
l0 = 0
# L = 4
T = 1
r = 0.05
var = 0.2
M = 100
mse = []
mse_as = []
mse_look = []
for L in range(4, 9):
  c = 0
  c_as = 0
  c_look = 0
  N = [((L -l0 +1) * np.power(2, 2 * L - i)) for i in range(l0, L + 1)]
  print(N)
  # df = np.zeros((M, 5))
  for k in range(M):
    value = 0
    value_as = 0
    value_look = 0
    for l in range(l0, L + 1):
      if l == l0:
        dl = 1/np.power(2, l)
        S = np.ones(N[l - l0])
        S_as = np.ones(N[l - l0])
        S_look = np.ones(N[l - l0])
        for i in range(T * np.power(2, l)):
          Z = np.random.normal(0, np.sqrt(dl), N[l - l0])
          S = S + r * dl * S + var * S * Z
          S_as += S
          S_look = np.minimum(S_look, S)
        S_as = S_as/(T * np.power(2, l) + 1) 
        P =  S - 1
        P[P < 0] = 0
        P = np.exp(r) * P
        value += np.sum(P)/N[l - l0]
        P_as =  S_as - 1
        P_as[P_as < 0] = 0
        P_as = np.exp(r) * P_as
        value_as += np.sum(P_as)/N[l - l0]
        P_look =  S_look - 0.5
        P_look[P_look < 0] = 0
        P_look = np.exp(r) * P_look
        value_look += np.sum(P_look)/N[l - l0]
        # print(np.sum(P)/N[l - l0])
      else:
        dl1 = 1/np.power(2, l)
        dl2 = 1/np.power(2, (l-1))    
        S1 = np.ones(N[l - l0])
        S2 = np.ones(N[l - l0])
        S1_as = np.ones(N[l - l0])
        S2_as = np.ones(N[l - l0])
        S1_look = np.ones(N[l - l0])
        S2_look = np.ones(N[l - l0])
        for i in range(T * np.power(2, (l-1))):
          Z1 = np.random.normal(0, np.sqrt(dl1), N[l - l0])
          S1 = S1 + r * dl1 * S1 + var * S1 * Z1
          Z2 = np.random.normal(0, np.sqrt(dl1), N[l - l0])
          S1 = S1 + r * dl1 * S1 + var * S1 * Z2
          S2 = S2 + r * dl2 * S2 + var * S2 * (Z1 + Z2)
          S1_as += S1
          S2_as += S2
          S1_look = np.minimum(S1, S1_look)
          S2_look = np.minimum(S2, S2_look)
        S1_as = S1_as/(T * np.power(2, l-1) + 1)
        S2_as = S2_as/(T * np.power(2, l-1) + 1) 
        P1 =  S1 - 1
        P1[P1 < 0] = 0
        P1 = np.exp(r) * P1
        P2 =  S2 - 1
        P2[P2 < 0] = 0
        P2 = np.exp(r) * P2
        value += np.sum(P1 - P2)/N[l - l0]
        P1_as =  S1_as - 1
        P1_as[P1_as < 0] = 0
        P1_as = np.exp(r) * P1_as
        P2_as =  S2_as - 1
        P2_as[P2_as < 0] = 0
        P2_as = np.exp(r) * P2_as
        value_as += np.sum(P1_as - P2_as)/N[l - l0]
        P1_look =  S1_look - 0.5
        P1_look[P1_look < 0] = 0
        P1_look = np.exp(r) * P1_look
        P2_look =  S2_look - 0.5
        P2_look[P2_look < 0] = 0
        P2_look = np.exp(r) * P2_look
        value_look += np.sum(P1_look - P2_look)/N[l - l0]
        # df[k, l] = np.sum(P1 - P2)/N[l - l0]
        # print(np.sum(P1)/N[l - l0], np.sum(P2)/N[l - l0])
    c += np.power(value - real, 2)
    c_as += np.power(value_as - real_as, 2)
    c_look += np.power(value_look - real_look, 2)
    print(value_look, k)
  mse.append(c/M)
  mse_as.append(c_as/M) 
  mse_look.append(c_look/M) 

# print(df) 
```
